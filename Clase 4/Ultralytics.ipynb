```python
from PIL import Image
from IPython.display import Image as DisplayImage
from ultralytics import YOLO
import ultralytics

# chequeos para asegurarse de que está todo en orden y listo para ser usado
ultralytics.checks()
```

```python
# Elegir el modelo preentrenado
# Más modelos en https://docs.ultralytics.com/models/yolov8/#supported-tasks-and-modes

model_name = 'yolov8n.pt'      # Para detección
# model_name = 'yolov8n-seg.pt'  # Para segmentación
# model_name = 'yolov8n-cls.pt'  # Para clasificación
# model_name = 'yolov8n-pose.pt'  # Para detección de poses

# Carga el modelo preentrenado elegido
model = YOLO(model_name)
```

```python
# La ruta y nombre del archivo con la imagen/video a analizar
# La imagen/video puede ser de cualquier resolución
# El modelo hace el pre-procesamiento necesario

image_file = "Test.jpg"
```

```python
results = model(image_file)  # Predicción
if len(results)==1:
    # La entrada es una imagen
    r = results[0]
    im_array = r.plot()
    im = Image.fromarray(im_array[..., ::-1])
    display(im)
else:
    # La entrada es un video
    total_frames = len(results)
    step = 50   # Mostrar un frame cada "step" frames
    i = 0
    while i < total_frames:
        r = results[i]
        im_array = r.plot()
        im = Image.fromarray(im_array[..., ::-1])
        print("Frame: " + str(i))
        display(im)
        i = i + step
```
